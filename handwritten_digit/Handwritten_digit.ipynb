### Preprocessing the Image
#### image pixel Below from 0 - 255
#### divide data by 255 to scale the values from 0-1

```python
xtrain_scl = xtrain/255
xtest_scl = xtest/255
```

```python
xtrain_scl[0]
```

```python
import numpy as np
```

```python
np.unique(ytrain)   # we have here multi class problem
```

```python
# it is similar as an one hot encoding
from keras.utils import  to_categorical
ytrain_cat=to_categorical(ytrain)
ytest_cat=to_categorical(ytest)
```

```python
ytrain_cat[0]
```

### Bulid the CNN (Convolution Neural Network)

```python
from keras.models import Sequential
from keras.layers import Convolution2D,MaxPool2D,BatchNormalization,Dropout,Flatten,Dense
from keras.callbacks import EarlyStopping
```

```python
model=Sequential()

# First Convolution layer
model.add(Convolution2D(filters=16,kernel_size=(3,3),input_shape=(28,28,1)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Dropout(0,2))

# second Convolution layer
model.add(Convolution2D(filters=16,kernel_size=(3,3)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Dropout(0,2))  

# flatten layer
model.add(Flatten())

# Dence layers
model.add(Dense(128,activation="relu"))
model.add(Dense(128,activation="relu"))

# Output layer
model.add(Dense(10,activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
early_stop=EarlyStopping(monitor="val_loss",patience=3)
```

```python
hist=model.fit(xtrain_scl,ytrain_cat,validation_split=0.2,epochs=100,callbacks=[early_stop])
```

### Plot the Learning Curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(hist.history["loss"],label="Training loss")
plt.plot(hist.history["val_loss"],label="validation loss")
plt.title("Learning Curve")
plt.xlabel("Epochs")
plt.ylabel("Categorical cross Entropy loss")
plt.show
```

### Evaluating The model

```python
model.evaluate(xtrain_scl,ytrain_cat)
```

```python
model.evaluate(xtest_scl,ytest_cat)
```
